### 1. Preprocessing

```python
# Load packages
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
# Display settings
pd.set_option('display.max_columns', None)  
pd.set_option('display.max_rows', None)  
pd.set_option('display.max_colwidth', None)
```

```python
# Read data
cluster = pd.read_csv('Data/Clustering/All_withcluster.csv')
prod1 = pd.read_csv('Data/Regression/Production.csv')
loss1 = pd.read_csv('Data/Regression/FoodProductionLoss.csv')
prod2 = pd.read_csv('Data/Regression/Production2.csv')
loss2 = pd.read_csv('Data/Regression/FoodProductionLoss2.csv')
```

```python
# Create list
c1 = cluster[cluster['Cluster']=='C1']
countries = c1['Country'].tolist()
crops = prod1['Item'].unique()
```

```python
crops
```

```python
# Create a dataframe with all combined data
df1 = pd.DataFrame({'Country': countries})
df2 = pd.DataFrame({'Country':countries})

# df1 (2016 - 2020) 
for country in countries:
    for crop in crops:
        value = prod1[(prod1['Country'] == country) & (prod1['Item'] == crop) & (prod1['Element'] == 'Production')]['Value']
        column_name = f'{crop}'
        if not value.empty:
            df1.loc[df1['Country'] == country, column_name] = value.mean()
        else:
            df1.loc[df1['Country'] == country, column_name] = 'No Data'

# df2 (2011 - 2015) 
for country in countries:
    for crop in crops:
        value = prod2[(prod2['Country'] == country) & (prod2['Item'] == crop) & (prod2['Element'] == 'Production')]['Value']
        column_name = f'{crop}'
        if not value.empty:
            df2.loc[df2['Country'] == country, column_name] = value.mean()
        else:
            df2.loc[df2['Country'] == country, column_name] = 'No Data'
```

```python
# Keep columns for specific crops
columns = ['Country','Maize (corn)','Rice','Sugar cane','Tomatoes', 'Bananas','Groundnuts, excluding shelled', 'Cassava, fresh','Cocoa beans','Coffee, green','Pineapples','Sweet potatoes', 'Mangoes, guavas and mangosteens', 'Coconuts, in shell' ]
df1 = df1[columns]
df2 = df2[columns]
```

```python
# Add Y (food loss)

# df1
for country in df1['Country']:
    value = loss1[(loss1['Country'] == country) & (loss1['Element'] == 'Loss')]['Value']
    column_name = 'Loss'
    if not value.empty:
        df1.loc[df1['Country'] == country, column_name] = value.mean()
    else:
        df1.loc[df1['Country'] == country, column_name] = 'No Data'
        
# df2 
for country in df2['Country']:
    value = loss2[(loss2['Country'] == country) & (loss2['Element'] == 'Loss')]['Value']
    column_name = 'Loss'
    if not value.empty:
        df2.loc[df2['Country'] == country, column_name] = value.mean()
    else:
        df2.loc[df2['Country'] == country, column_name] = 'No Data'
```

```python
# Treat 'No Data' as NaN
df1.replace('No Data', np.nan, inplace=True)
df2.replace('No Data', np.nan, inplace=True)
```

```python
df1 = df1[df1['Country']!='Equatorial Guinea']
df1 = df1[df1['Country']!='Eritrea']

df2 = df2[df2['Country']!='Equatorial Guinea']
df2 = df2[df2['Country']!='Eritrea']
```

```python
# Save data
df1.to_csv('Data/Regression/All1.csv',index=False)
df2.to_csv('Data/Regression/All2.csv',index=False)
```
